```python
import pandas as pd
import ast
dataset = pd.read_csv("/Users/komlan/JungleBike/datasets/scrapp_data.csv")
dataset = dataset.rename(columns={'Unnamed: 0':'product_id'})

def get_value(row, key):
    try:
        value = row[key]
        return value if value != '' else None
    except:
        return None
    
dataset.views = dataset.views.apply(lambda x: ast.literal_eval(x))
dataset = dataset.explode('views')
dataset['user'] = dataset.views.apply(lambda row: get_value(row, 'name'))
dataset['city'] = dataset.views.apply(lambda row: get_value(row, 'city'))
dataset['age'] = dataset.views.apply(lambda row: get_value(row, 'Âge'))
dataset['activity'] = dataset.views.apply(lambda row: get_value(row, 'Quel sportif êtes-vous ?'))
dataset['level'] = dataset.views.apply(lambda row: get_value(row, 'Quel est votre niveau ?'))
dataset['vote'] = dataset.views.apply(lambda row: get_value(row, 'rate'))
dataset.vote = dataset.vote.str.extract('([0-5])')
dataset.age = dataset.age.apply(lambda x: x.replace(' à ','-').replace('ans','') if not pd.isnull(x) else None)
dataset['avis'] = dataset.views.apply(lambda row: get_value(row, 'remark'))
dataset['description'] = dataset.views.apply(lambda row: get_value(row, 'description'))
dataset = dataset.drop(['views','link'], axis=1)
dataset = dataset.reset_index(drop=True)
dataset
```

```python
import pandas as pd
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager


def get_value_xpath(element, selector):
    try:
        return element.find_element_by_xpath(selector).text
    except:
        return None
    
def get_elements(element, selector):
    try:
        return element.find_elements_by_class_name(selector)
    except:
        return None
    
def get_value(element, selector):
    try:
        return element.find_element_by_class_name(selector).text
    except:
        return None
    
def get_users_views(users_infos_pr):
    product_users_infos = []
    for i in range(len(users_infos_pr)):
        element = users_infos_pr[i]
        user = {}
        user['name'] = get_value_xpath(element, ".//button[@class='bv-author bv-fullprofile-popup-target bv-focusable']")
        user['city'] = get_value(element, 'bv-author-location')
        
        users_infos = get_elements(element, 'bv-author-cdv')
        
        for info in users_infos:
            key = get_value(info, 'bv-author-userinfo-data')
            value = get_value(info, 'bv-author-userinfo-value')
            user[key] = value
        
        user['rate'] = get_value(element, 'bv-off-screen')
        user['remark'] = get_value(element, 'bv-content-title')
        user['description'] = get_value(element, "bv-content-summary-body-text")

        product_users_infos.append(user)
    return product_users_infos


browser = webdriver.Firefox(executable_path=GeckoDriverManager().install())
browser.implicitly_wait(5) # seconds


all_products = []

links = ['https://www.alltricks.fr/F-183706-pneus-route/P-532533-pneu_route_continental_gp_5000_700_mm_tubetype_souple_vectran_breaker_blackchili',
         'https://www.alltricks.fr/F-11911-cassettes/P-259868-cassette_shimano_ultegra_cs_r8000_11v']

for i, url in enumerate(links):
    try:
        browser.get(url)
        all_users_infos_pr = browser.find_element_by_xpath("//ol[@class='bv-content-list bv-content-list-reviews']")
        product_name = get_value(browser, 'heading-title-text')
        brand = get_value(browser, 'heading-line-link-brand')
        users_infos_pr = all_users_infos_pr.find_elements_by_class_name('bv-content-top-review')
        views = get_users_views(users_infos_pr)
        all_products.append([product_name, brand, views, url])
        print("good ",i)
    except Exception as e:
        print("bad ",e)
        bad_url.append(url)


#Save scraped data
data = pd.DataFrame(all_products, columns = ["product_name","brand","views", "link"])
data
```
