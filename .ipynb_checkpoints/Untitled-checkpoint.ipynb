```python
from datetime import datetime
today = datetime.today().strftime('%Y/%m/%d')

def transform_table(table):
    table = table.reset_index(drop=True)
    columns = table.loc[2].values[2:4].tolist()+table.loc[0].values[4:].tolist()
    table = table.drop(['Unnamed: 0','Unnamed: 1'], axis=1)
    table.columns = columns
    table = table[3:]
    table['Entity'] = None
    entity_columns = table.columns[3:-1]

    for i, row in table.iterrows():
        curr_val = []
        for column in entity_columns:
             curr_val.append([column, row[column]])

        table.at[i, 'Entity'] = curr_val
    table = table.drop(entity_columns, axis=1)
    table = table.explode('Entity')
    table['Value'] = table.Entity.apply(lambda x: x[1])
    table.Entity = table.Entity.apply(lambda x: x[0])
    table['Period'] = table.columns[2]
    table['Update'] = today
    table = table.rename(columns={table.columns[2]: 'Unit'})
    table = table[['Update','Period','Code','Label','Entity','Unit','Value']]
    return table

data_1_t = transform_table(data[5:31])
data_2_t = transform_table(data[33:58])
data_3_t = transform_table(data[61:87])
data_4_t = transform_table(data[33:58])
data_5_t = transform_table(data[61:87])
data_6_t = transform_table(data[89:115])
data_7_t = transform_table(data[117:143])
data_8_t = transform_table(data[145:171])
data_t = pd.concat([data_1_t, data_2_t, data_3_t, data_4_t, data_5_t, data_6_t, data_7_t, data_8_t])
data_t
```

```python
data_t.Value = data_t.Value.apply(lambda x: float(str(x).replace('-','0'))) 
table = pd.pivot_table(data_t, values='Value', index=['Code','Entity'],
                    columns=['Period'], aggfunc=np.sum).reset_index()

dic = {'POCORA' : 'Argentina', 'POCOBE' : 'France',
      'TOPBRACO' : 'Germany','PORBCO' : 'Mexico','NOCOCPO' : 'Japan','POSECO' : 'China'}
table.Entity = table.Entity.apply(lambda x: dic[x])
table.to_excel('/Users/komlan/Downloads/output.xlsx')
```

```python


def computeJoinPoint(sequence1, sequence2):
    if(sequence1 >= MAX or sequence2 >= MAX):
        return -1

    if(sequence1 == sequence2):
        return sequence1

    next1 = sequence1 + sum([int(el) for el in list(str(sequence1))])
    next2 = sequence2 + sum([int(el) for el in list(str(sequence2))])

    if next1 == sequence2:
        return next1

    if(next2 == sequence1):
        return next2

    if(next1 == next2):
        return next1

    if(next1 < sequence2):
        return computeJoinPoint(next1, sequence2)

    if(next2 < sequence1):
        return computeJoinPoint(next2, sequence1)

    return computeJoinPoint(next1, next2)

MAX = 10000000;
sequence1 = 471
sequence2 = 480

print("The join point of ", sequence1 ," and ", sequence2,
         " is : ", computeJoinPoint(sequence1, sequence2));
```
