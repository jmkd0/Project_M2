```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)#np.nan
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
from sklearn import tree
import datetime
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, r2_score

raw_data = pd.read_csv("Data/credit_scoring.csv", sep=';')
raw_data.head(15)
```

```python
#np.nan with sys.maxsize
raw_data.info()
```

```python
raw_data.describe()
```

```python
target = raw_data["Status"].values
data_num = raw_data.drop(["Status"], axis=1).values
data_num
```

```python
sns.countplot(raw_data["Status"])
```

```python
#Spliting data to training and test set
X_train, X_test, y_train, y_test = train_test_split(data_num, target, train_size=0.8, test_size=0.2, random_state=42)
```

## Data Exploration: Apply models on the data

```python
#Decision Tree
dt_clf = DecisionTreeClassifier(max_depth=5, criterion='entropy', random_state=12)
dt_clf.fit(X_train, y_train)
predicted = dt_clf.predict(X_test)

print("Score ", accuracy_score(y_test, predicted))
print(classification_report(y_test, predicted))
```

```python
#K - Nearest Neighbor
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_train, y_train)
predicted = knn_clf.predict(X_test)
print("Score ", accuracy_score(y_test, predicted))
print(classification_report(y_test, predicted))
```

## Data Normalisation

### 1. Scalling Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.fit_transform(X_test)
```

```python
#Decision tree on standard Scaling data
dt_clf_s = DecisionTreeClassifier(max_depth=5, criterion='entropy', random_state=0)
dt_clf_s.fit(X_train_s, y_train)
predicted = dt_clf_s.predict(X_test_s)
print("Score ", accuracy_score(y_test, predicted))
print(confusion_matrix(y_test, predicted))
print(classification_report(y_test, predicted))
```

```python
#K-Nearest Neighbor on standard Scaling data
knn_clf_s = KNeighborsClassifier(n_neighbors=5)
knn_clf_s.fit(X_train_s, y_train)
predicted = knn_clf_s.predict(X_test_s)
print("Score ", accuracy_score(y_test, predicted))
print(classification_report(y_test, predicted))
```

### 2. MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_m = scaler.fit_transform(X_train)
X_test_m = scaler.fit_transform(X_test)
```

```python
#Decision tree on MinMax Scaling data
dt_clf_m = DecisionTreeClassifier(max_depth=5, random_state=12)
dt_clf_m.fit(X_train_m, y_train)
predicted = dt_clf_m.predict(X_test_m)
print("Score ", accuracy_score(y_test, predicted))
print(classification_report(y_test, predicted))
```

```python
#K-Nearest Neighbor on MinMax Scaling data
knn_clf_m = KNeighborsClassifier(n_neighbors=5)
knn_clf_m.fit(X_train_m, y_train)
predicted = knn_clf_m.predict(X_test_m)
print("Score ", accuracy_score(y_test, predicted))
print(classification_report(y_test, predicted))
```

Decision Tree:
On remarque que le modele entrain√© sur les donn√©es normalis√© n'a pas √©t√© plus performant par rapport √† celui entrain√© sans normalisation. On conclu que le decision tree n'est pas aussi efficasse sur la donn√©e normalis√©e.

KNN:
Dans ce cas on a au contraire une am√©lioration du score avec la data normalis√©e par rapport √† la data non normalis√©e.

## PCA with Scalling Data 

```python
from sklearn.decomposition import PCA
k = 3

scaler = StandardScaler()
X_train_1 = scaler.fit_transform(X_train)
X_train_pca = PCA(n_components=k).fit_transform(X_train_1)
X_train_pca_norm = np.concatenate((X_train_1, X_train_pca), axis=1)

X_test_1 = scaler.fit_transform(X_test)
X_test_pca = PCA(n_components=k).fit_transform(X_test_1)
X_test_pca_norm = np.concatenate((X_test_1, X_test_pca), axis=1)
```

```python
#Decision tree on standard Scaling data
dt_clf_pca = DecisionTreeClassifier(max_depth=5, criterion='entropy', random_state=0)
dt_clf_pca.fit(X_train_pca_norm, y_train)
predicted = dt_clf_pca.predict(X_test_pca_norm)
print("Score ", accuracy_score(y_test, predicted))
print(classification_report(y_test, predicted))
```

```python
#K-Nearest Neighbor on standard Scaling data
knn_clf_pca = KNeighborsClassifier(n_neighbors=5)
knn_clf_pca.fit(X_train_s, y_train)
predicted = knn_clf_pca.predict(X_test_s)
print("Score ", accuracy_score(y_test, predicted))
print(classification_report(y_test, predicted))
```
