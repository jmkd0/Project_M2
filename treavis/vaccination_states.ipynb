```r
raw_data = read.csv(file = 'data.csv')
head(raw_data)
```

```r
genre = c("0"="Homme", "1"="Femme")
region = c("0"="Auvergne-Rhône-Alpes", "1" = "Bourgogne-Franche-Comté", "2" = "Bretagne", "3" = "Centre-Val de Loire", "4" = "Corse",
            "5" = "Grand Est", "6" = "Hauts de France", "7" = "Ile-de-France", "8" = "Normandie", "9" = "Nouvelle Aquitaine", 
           "10" = "Occitanie", "11" = "Pays de la Loire", "12" = "Provence-Alpes-Côte d'Azur")
diplome = c("0"= "Pas de diplôme", "1"= "CEP / Brevet des collèges", "2"= "BEP / CAP", "3"="Baccalauréat", "4"= "Bac +1",
             "5"= "Bac +2", "6"= "Bac +3", "7"= "Bac +4", "8"= "Bac +5", "9"= "Bac +8")
vaccination = c("0"= "Non", "1" = "Une dose", "2"= "Deux doses")
data = raw_data
transform = function(vector, values){
    vec = c()
    for (v in vector)
        vec = c(vec, values[toString(v)])
    vec
}
data[,"genre"] = transform(data$genre, genre)
data[,"region"] = transform(data$region, region)
data[,"diplome"] = transform(data$diplome, diplome)
data[,"vaccination"] = transform(data$vaccination, vaccination)
head(data)
```

# Etude de la variable diplome

```r
library(ggplot2)
diplomes = aggregate(data$identifiant, by=list(data$diplome), FUN=length)
# plotting a barplot with counts
options(repr.plot.width = 15, repr.plot.height = 8) 
ggplot(diplomes, aes(x = Group.1, y = x, fill = Group.1)) + 
  geom_bar(stat = "identity") +
  geom_text(aes(label = x), vjust = 0)
```

```r
library("dplyr")
age_tranche = function(vector){
    ages_tr = c(18, 25, 35, 50, 70, 100)
    len = length(ages_tr)
    vec = c()
    for (v in vector){
        inter = ""
        for(i in c(1:len)){
            if (i < len && between(v, ages_tr[i], ages_tr[i+1])){
                inf = paste("(", toString(ages_tr[i]), sep='')
                sup = paste(toString(ages_tr[i+1]),"]", sep='')
                inter = paste(inf, sup, sep=',')
                
                break
            }
        }
        vec <- c(vec, inter)
    }
    vec
}

data[,"age"] = 2022 - as.integer(data$date_naissance)
data[,"age_cat"] = age_tranche(data$age)
head(data)
```

# 4.

```r
vac_conf = table(data[,"vaccination"] , data[,"confinement_respect"])
vac_conf = cbind(vac_conf, rowSums(vac_conf))
vac_conf = rbind(vac_conf, colSums(vac_conf))
row.names(vac_conf)[4] = "Total"
colnames(vac_conf)[12] = "Total"
vac_conf
```

```r
vac_conf_freq = matrix(nrow=4, ncol=12)
for (i in 1:nrow(vac_conf))
    vac_conf_freq[i,] = round((vac_conf[i,]/vac_conf[i,12])*100, 2)
row.names(vac_conf_freq) = row.names(vac_conf)
colnames(vac_conf_freq) = colnames(vac_conf)
vac_conf_freq
```

```r
vac_conf_pl = subset(data, select = c("vaccination", "confinement_respect"))
ggplot(vac_conf_pl, aes(fill=vaccination, x=confinement_respect)) + 
    geom_bar(position="fill")
```
