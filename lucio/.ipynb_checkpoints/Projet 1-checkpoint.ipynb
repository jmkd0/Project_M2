## BAYES model

```python
df = shuffle(df)
df.head(10)
```

```python
fig, ax = plt.subplots(figsize=(14, 7), subplot_kw=dict(aspect="equal"))

recipe = ["10920 Adware_(20,5%)",
          "1133 Backdoor_(2,1%)",
          "248 FileInfector_(0,5%)",
          "14053 Riskware_(26,3%)",
          "1932 No_Category_(3,6%)",
          "1290 PUA_(2,4%)",
          "3411 Ransomware_(6,4%)",
          "886 Scareware_(1,7%)",
          "8437 Trojan_(15,7%)",
          "241 Trojan_Banker_(0,5%)",
          "1570 Trojan_Dropper_(2,9%)",
          "1939 Trojan_SMS_(3,6%)",
          "2840 Trojan_Spy_(5,3%)",
          "4475 Zero_Day_(8,4%)"]

data = [float(x.split()[0]) for x in recipe]
ingredients = [x.split()[-1] for x in recipe]


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, ingredients,
          title="Categories",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=10, weight="bold")

ax.set_title("Répartition des catégories")

plt.show()
```

```python
Nombre_Zero_Day = len(df[df.Category == 'Zero_Day'])

print(f"Nombre Zero_Day: {Nombre_Zero_Day}")
```

```python
mapping ={'Adware':0,'Backdoor':1,'FileInfector':2,'Riskware':3,'No_Category':4,'PUA':5,'Ransomware':6,
'Scareware':7,'Trojan':8,'Trojan_Banker':9,'Trojan_Dropper':10,'Trojan_SMS':11,'Trojan_Spy':12,'Zero_Day':13}

df = df.replace({'Category':mapping})
```

```python
x = df.drop(['Category','Family','state'],axis=1)
y = df['Category']
```

```python
df.head(10)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state = 1)

print(f"x_train: {x_train.shape}")
print(f"x_test: {x_test.shape}")
print(f"y_train: {y_train.shape}")
print(f"y_test: {y_test.shape}")
```

```python
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
y_pred = nb.predict(x_test)

cm = confusion_matrix(y_test,y_pred)

print(cm)
print()
print('accuracy: ',accuracy_score(y_test,y_pred))
```
